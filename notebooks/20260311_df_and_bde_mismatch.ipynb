```python
temps = pd.read_fwf("/home/bekah/m3_papers/20260310/m3_detector_temperature.tab")
```

```python
temps
```

```python
# want to get temperature for every flat, bde, and dark signal image vs the observation 
# temp to see how similar all the temperatures are. if there's an observation where 
# they are significantly different, that seems like a problem. 

temp_mapping = temps.set_index('id')['temp']
df['obs_temp'] = df['id'].map(temp_mapping)
df['flat_temp'] = df['flat'].map(temp_mapping)
df['dark_temp'] = df['dark'].map(temp_mapping)
df['bde_temp'] = df['bde'].map(temp_mapping)
```

```python
df['max_diff'] = df['obs_temp'] - df['flat_temp']
df['max_diff'].describe()
```

```python
df['max_diff_dark'] = df['obs_temp'] - df['dark_temp']
df['max_diff_dark'].describe()
```

```python
df['max_diff_bde'] = df['obs_temp'] - df['bde_temp']
df['max_diff_bde'].describe()
```

```python
df['max_diff_flat_dark'] = df['flat_temp'] - df['dark_temp']
df['max_diff_flat_dark'].describe()
```
