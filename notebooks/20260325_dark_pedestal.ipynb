```python
# For every fits file in the folder I want to open it and calculate an array (1 for each file) that contains
# for each frame along NAXIS 3, for each row, I want the average pixel value for the middle 200 pixels, 
# and the individual values of pixels 1-4 and pixels 317-320. 
```

```python
import os
import glob
import numpy as np
from astropy.io import fits

FOLDER = "/home/bekah/m3_papers/l0_obs"

# Need to double check these pixel ranges vs DPSIS vs what they are in real life 
EDGE_LOW  = slice(0, 4)     
EDGE_HIGH = slice(316, 320)  

fits_files = sorted(
    glob.glob(os.path.join(FOLDER, "*.fits")) +
    glob.glob(os.path.join(FOLDER, "*.fit"))  +
    glob.glob(os.path.join(FOLDER, "*.fts"))
)
print(f"Found {len(fits_files)} FITS file(s)")

results = {} 

for path in fits_files:
    fname = os.path.basename(path)
    with fits.open(path, memmap=True) as hdul:
        data = hdul[0].data
        if data is None:
            for hdu in hdul[1:]:
                if hdu.data is not None and hdu.data.ndim >= 2:
                    data = hdu.data
                    break
        data = np.array(data, dtype=float)
    if data.ndim == 2:
        data = data[np.newaxis, ...]

    n_frames, n_rows, n_cols = data.shape
    mid_slice = slice(n_cols // 2 - 100, n_cols // 2 + 100)

    out = np.empty((n_frames, n_rows, 9), dtype=float)
    out[:, :, 0]   = data[:, :, mid_slice].min(axis=2)
    out[:, :, 1:5] = data[:, :, EDGE_LOW]
    out[:, :, 5:9] = data[:, :, EDGE_HIGH]

    results[fname] = out
    print(f"  {fname}: shape={out.shape}")
```
