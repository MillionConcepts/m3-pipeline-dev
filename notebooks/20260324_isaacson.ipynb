```python
import pandas as pd 
```

```python
op1b = pd.read_csv('op1b_search.csv', header=None)[4]
op2c = pd.read_csv('op2c1_searchresults.csv', header=None)[4]
temps = pd.read_fwf("/home/bekah/m3_papers/20260310_darkcurrent/m3_detector_temperature.tab")
```

```python
temps
```

```python
def parse_col(series, source_name):
    cleaned = series.str.strip()
    df = pd.DataFrame({
        'id': cleaned.str.extract(r'(M3G\d+T\d+)')[0],
        'version': cleaned.str.extract(r'_(V\d+)_')[0],
        'source': source_name
    })
    return df

result = pd.concat([
    parse_col(op1b, 'op1b'),
    parse_col(op2c, 'op2c')
], ignore_index=True)

print(result)
```

```python
result
```

```python
result = result.merge(temps, on='id', how='left')
```

```python
hdr_info = pd.read_csv("/home/bekah/m3_papers/20260310_darkcurrent/l1b_rdn_hdr_info.csv")
result = pd.merge(result, hdr_info, on=['id', 'version'], how='left')
```

```python
result['dark_id'] = result['cal_dark_signal_image'].str.split('.').str[0]
```

```python
result['dark_id'] = result['dark_id'].str.split('_').str[0]
```

```python
result = result.merge(temps, left_on='dark_id', right_on='id', how='left')
```
